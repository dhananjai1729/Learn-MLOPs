```python
import mlflow
```

```python
# set tracking uri
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
# set an experiment
mlflow.set_experiment("testing experiment")


# create a run 
with mlflow.start_run() as run:
    # log parameters
    mlflow.log_param("param1", 1)
    mlflow.log_param("param2", "test")

    # log metrics
    mlflow.log_metric("accuracy", 0.98)
    mlflow.log_metric("loss", 0.02)
```
